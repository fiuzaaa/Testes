## Modelo Preditivo 

### Crescimento da carteira para os próximos 2 anos em termos de volume de operações e da carteira

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


df = pd.DataFrame(df_mensal)
df = df.reset_index(drop=True)


# Divisão entre treino e teste
X = df.index.values.reshape(-1, 1)  # Variável independente (meses)
y = df['valor_principal'].values  # Variável dependente

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Treinamento do modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Previsão para os próximos 2 anos
anos_futuros = np.arange(1, 71, 1).reshape(-1, 1)  # 24 meses no futuro
previsao = model.predict(anos_futuros)

# Cálculo do RMSE
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")

# Limites superior e inferior da faixa de erro
limite_superior = previsao + rmse
limite_inferior = previsao - rmse

# Plotando os dados
plt.figure(figsize=(10, 6))  # Tamanho da figura
plt.scatter(X_train, y_train, color='blue', label='Dados de Treino')  # Pontos de treino
plt.scatter(X_test, y_test, color='red', label='Dados de Teste')  # Pontos de teste
plt.plot(anos_futuros, previsao, color='green', label='Previsão')  # Linha de previsão
plt.fill_between(anos_futuros.flatten(), limite_inferior, limite_superior, color='gray', alpha=0.3, label='Faixa de Erro')
plt.xlabel('Meses')
plt.ylabel('Valor Principal')
plt.title('Previsão de Valor Principal ao Longo dos Meses com Faixa de Erro')
plt.legend()
plt.grid(True)
plt.show()
```

### Valor encontrado para previsão daqui 2 ano com a regressao linear mensal

```python
valor_previsto_linear = round(previsao[-1],2)
print('Projeção linear para 2 anos:', valor_previsto_linear)
print('Projeção linear para 2 anos desvio sup:', round(limite_superior[-1],2))
print('Projeção linear para 2 anos desvio inf:', round(limite_inferior[-1],2))
```
